# Regression Trees & Random Forest Regression

## Libraries and settings

```python
# Libraries
import os
import numpy as np
import pandas as pd
from sklearn import tree
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
from sklearn.datasets import make_regression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

# Ignore warnings
import warnings
warnings.filterwarnings('ignore')

# Show current working directory
print(os.getcwd())
```

## Import the apartment data

```python
# Define columns for import
columns = [ 'web-scraper-order',
            'address_raw',
            'rooms',
            'area',
            'luxurious',
            'price',
            'price_per_m2',
            'lat',
            'lon',
            'bfs_number',
            'bfs_name',
            'pop',
            'pop_dens',
            'frg_pct',
            'emp',
            'mean_taxable_income',
            'dist_supermarket']

# Read and select variables
df_orig = pd.read_csv("apartments_data_enriched_cleaned.csv", sep=";", encoding='utf-8')[columns]

# Rename variable 'web-scraper-order' to 'apmt_id'
df_orig = df_orig.rename(columns={'web-scraper-order': 'id'})

# Remove missing values
df = df_orig.dropna()
df.head(5)

# Remove duplicates
df = df.drop_duplicates()

# Remove some 'extreme' values
df = df.loc[(df['price'] >= 1000) & 
            (df['price'] <= 5000)]

# Reset index
df = df.reset_index(drop=True)

print(df.shape)
df.head(5)
```

## Regression Tree
See also: https://data36.com/regression-tree-python-scikit-learn

### Create train and test samples for the regression tree (train = 80%, test = 20% of the data)

```python
# Create train and test samples
X_train, X_test, y_train, y_test = train_test_split(df[['area', 
                                                        'rooms',
                                                        'pop_dens',
                                                        'mean_taxable_income',
                                                        'dist_supermarket']], 
                                                        df['price'], 
                                                        test_size=0.20, 
                                                        random_state=42)

# Show X_train
print('X_train:')
print(X_train.head(), '\n')

# Show y_train
print('y_train:')
print(y_train.head())
```

### Fit the regression tree model

```python
# Create decision tree regressor object
reg = DecisionTreeRegressor(random_state=20, max_depth=3)
reg5 = DecisionTreeRegressor(random_state=21, max_depth=5)

# Train decision tree regressor
reg = reg.fit(X_train, y_train)
reg5 = reg5.fit(X_train, y_train)
# Predict the response for test dataset
y_pred = reg.predict(X_test)
y_pred5 = reg5.predict(X_test)
```

### Calculate coefficient of determination (R-squared)

```python
# Calculate coefficient of determination, round to 4 decimals
print(f'R-squared max depth 3: {r2_score(y_test, y_pred):.4f}')
print(f'R-squared max depth 5: {r2_score(y_test, y_pred5):.4f}')
```

### Print text representation of the regression tree

```python
# Text representation of the regression tree
text_representation = tree.export_text(reg, 
                                       feature_names=list(X_train.columns))

# Print text_representation
print(text_representation)
```

### Vizualizing the regression tree

```python
fig = plt.figure(figsize=(12,6))
_ = tree.plot_tree(reg, 
                   feature_names=list(X_train.columns),  
                   class_names=['price'],
                   filled=True,
                   fontsize=9,
                   label='root',
                   rounded=True)
```

```python
fig = plt.figure(figsize=(12, 6))
_ = tree.plot_tree(reg5, 
                   feature_names= list(X_train.columns),
                   class_names=['price'],
                   filled= True,
                   fontsize= 9,
                   label= 'root',
                   rounded= True)
```

### Result comparison
The new regression tree has a better R-squared score. This is because, since it has more possible outcomes due to it's bigger depth, it has more specific intervals and thus a slightly lower error tendency. It also shows the difference in the variables taken into consideration for each subdivision of the tree and for each level, reflecting the information given by the feature importance analysis. 

## Removing area and rooms

```python
# Create train and test samples removing area and rooms
X_train, X_test, y_train, y_test = train_test_split(df[[#'area', 
                                                        #'rooms',
                                                        'pop_dens',
                                                        'mean_taxable_income',
                                                        'dist_supermarket']], 
                                                        df['price'], 
                                                        test_size=0.20, 
                                                        random_state=42)
```

```python
# Create decision tree regressor object
reg = DecisionTreeRegressor(random_state=20, max_depth=3)

# Train decision tree regressor
reg = reg.fit(X_train, y_train)

# Predict the response for test dataset
y_pred = reg.predict(X_test)
```

```python
print(f'R-squared max depth 3: {r2_score(y_test, y_pred):.4f}')
```

```python
fig = plt.figure(figsize=(12,6))
_ = tree.plot_tree(reg, 
                   feature_names=list(X_train.columns),  
                   class_names=['price'],
                   filled=True,
                   fontsize=9,
                   label='root',
                   rounded=True)
```

### Result

The new tree has an awful R-squared score. This is because, according to the feature importance chart area was the most important variable for the model; holding aprox. 56% of the information used by the model. So by removing this vital variable from the analysis, the model turns out completely useless.

## Random Forest Regression
For details see: https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestRegressor.html

### Create train and test samples for the random forest (train = 80%, test = 20% of the data)

```python
# Create train and test samples (the names X2_ and y2_ were used because X_ and y_ were already used above)
X2_train, X2_test, y2_train, y2_test = train_test_split(df[['area', 
                                                            'rooms',
                                                            'pop_dens',
                                                            'mean_taxable_income',
                                                            'dist_supermarket']], 
                                                            df['price'], 
                                                            test_size=0.20, 
                                                            random_state=42)

# Show X2_train
print('X2_train:')
print(X2_train.head(), '\n')

# Show y2_train
print('y2_train:')
print(y2_train.head())
```

### Fit the Random Forest Regression

```python
X, y = make_regression(n_features=4, n_informative=2,
                       random_state=5, shuffle=False)


reg_rf = RandomForestRegressor(n_estimators=500, 
                               max_depth=10, 
                               random_state=5)
reg_rf.fit(X2_train, y2_train)

# Calculate coefficient of determination (R-squared)
print(f'R-squared: {reg_rf.score(X2_test, y2_test):.4f}')
```

### Show feature importance

```python
cols = X2_train.columns

# Derive feature importance from random forest
importances = reg_rf.feature_importances_
std         = np.std([tree.feature_importances_ for tree in reg_rf.estimators_], axis=0)
indices     = np.argsort(importances)[::-1]

# Print col-names and importances-values
print( cols[indices] )
print( importances[indices] )

# Barplot with feature importance
df_fi = pd.DataFrame({'features':cols,'importances': importances})
df_fi.sort_values('importances', inplace=True)
df_fi.plot(kind='barh', 
           y='importances', 
           x='features', 
           color='darkred', 
           figsize=(6,3))

plt.show()
```

### Removing area result
The next variable to take the most importance from the model is rooms, with a 39.7% of the information retrieved by the model comming from this variable. However, the R-Squared value drops dramatically when the area variable is removed.
